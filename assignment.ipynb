```python
import pandas as pd
import numpy as np

df = pd.read_csv("country_vaccinations.csv")
df2 = pd.read_csv("country_vaccinations_by_manufacturer.csv")
print(df['date'].dtype)

df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['country', 'date'])
print(df['date'].dtype)

df.head()
```

### Analysis
The dataset is successfully loaded into Pandas DataFrames using the read_csv() function. The 'date' column is initially  as an object, so it is converted into datetime format to  proper time-based analysis because vaccination should be analysied over time . The data is then sorted by 'country' and 'date' to maintain chronological order(earliest to latest) within each country. Then the first few rows are displayed using head() to verify that the dataset is correctly loaded.

## Q1: Total vaccinations per country and disparity

```python
tot_vac = df.groupby('country')['total_vaccinations'].max()
top10 = tot_vac.sort_values(ascending=False).head(10)
bottom10 = tot_vac.sort_values().head(10)
print("Top 10 Countries:\n", top10)
print("\nBottom 10 Countries:\n", bottom10)
```

First , we group country by rows and select total vaccinations column and as there will be many records for each country over time we use max function take highest vaccination of each country. Then top 10 for selected using sort_values() which gives from highest to lowest by ascending = false and we want only 10 so head(10) we result first 10 and similarly for bottom 10 but lowest to highest. 
The results show a clear disparity in vaccination coverage, where highly populated and developed countries dominate the top positions, while smaller or less-developed countries appear in the lower range. This highlights inequality in vaccine distribution across the world.

Q2

```python
data = df[['people_vaccinated', 'people_fully_vaccinated']].dropna()
data['log_vacc'] = np.log(data['people_vaccinated'])
data['sqrt_ful'] = np.sqrt(data['people_fully_vaccinated'])
cor = data['people_vaccinated'].corr(data['people_fully_vaccinated'])

print("Correlation between vaccinated and fully vaccinated:", cor)

data.head()
```

The relationship between people vaccinated and full vaccinated was analyzed using NumPy ufuncs such logarithmic and square root transformations. Ufuncs are useful for normalizing data and reducing large differences in scales. The calculated correlation value was high and positive which indicates a strong relationship between the variables. The relationship was not proportionate because there is usually a time lag between the first and second dose. This means that a higher number vaccinated don't exactly have the same high number as full vaccinated, but there is a positive relationship between the two.

Q3

```python
df_h = df.set_index(['country', 'date'])

weekly = df_h.groupby(level=0).resample('W', level=1).sum(numeric_only=True)

IN = weekly.xs('India', level=0)
US = weekly.xs('United States', level=0)
BR= weekly.xs('Brazil', level=0)

IN_tab = IN[['total_vaccinations',
                     'people_vaccinated',
                     'people_fully_vaccinated']].reset_index()

US_tab= US[['total_vaccinations',
                 'people_vaccinated',
                 'people_fully_vaccinated']].reset_index()

BR_tab = BR[['total_vaccinations',
                       'people_vaccinated',
                       'people_fully_vaccinated']].reset_index()

print("India Table:\n", IN_tab.head())
print("\nUSA Table:\n", US_tab.head())
print("\nBrazil Table:\n", BR_tab.head())
```

### Analysis
The weekly data shows that vaccinations are increasing over time in India, the USA, and Brazil. The USA started earlier and has faster growth, while India and Brazil show slower and gradual increases.
At the beginning, the number of fully vaccinated people is very low because the second dose takes time. Later, it starts increasing.
Overall, all countries show an upward trend, but the speed of vaccination is different.

Q4

```python
df['eff_score'] = df['people_fully_vaccinated'] / df['people_vaccinated']

eff = df.groupby('country')['eff_score'].mean()
ranked = eff.sort_values(ascending=False)

res= ranked.to_frame(name='Efficiency Score')

print(res.head(10))
```

### Analysis
The efficiency score shows how many people completed both doses out of those who got the first dose. The data is grouped by country and ranked to find which countries are more efficient. Countries with higher scores complete vaccination better, while lower scores show incomplete vaccination.

Q5

```python
miss_per = df.isnull().mean() * 100

missing = miss_per[miss_per> 30]

print("Columns with >30% missing values:\n", missing)

ffill = df.ffill()
print("\nAfter forward fill:\n",
      ffill[['country','date','total_vaccinations',
                'people_vaccinated','people_fully_vaccinated']].head())
```

### Analysis
The percentage of missing values is calculated for each column, and several columns are found to have more than 30% missing data. Dropping these columns would lead to loss of important vaccination information, so imputation is preferred.

Forward-fill (ffill) is applied to fill missing values using the previous available data. This approach is suitable for time-series data because vaccination values change gradually over time, and the previous value provides a reasonable estimate. The output shows that missing values are filled while maintaining the continuity of the dataset. However, some initial rows may still contain missing values due to the absence of prior data.

```python
oxford_data = df[df['vaccines'].str.contains('Oxford/AstraZeneca', na=False)]

series = oxford_data.groupby('date')['total_vaccinations'].mean()

global_avg = df.groupby('date')['total_vaccinations'].mean()

compare = series - global_avg

print(compare.head())
```

### Analysis
The dataset is filtered to include only countries using the Oxford/AstraZeneca vaccine. The average total vaccinations for these countries are calculated for each date and compared with the global average. Subtraction is used to measure the difference between the two values.

The results show whether these countries perform above or below the global average. Positive values indicate higher vaccination levels, while negative values indicate lower performance. Some initial values appear as NaN due to missing data for certain dates. This happens because of index alignment, where Pandas matches values based on dates and produces NaN when data is not available in one of the Series.
Overall, this analysis demonstrates how vaccination performance of specific countries can be compared with global trends using index-aligned operations.

Q7

```python


df['daily_rate'] = df.groupby('country')['total_vaccinations'].diff()

df['daily_rate'] = df['daily_rate'].apply(lambda x: x if x > 0 else 0)

r = df[['country', 'date', 'daily_rate']]

r = r.set_index(['country', 'date'])

top_c = df.groupby('country')['daily_rate'].mean().sort_values(ascending=False)

print(r.head(10))
print("\nTop Performing Countries:\n", top_c.head(10))
```

### Analysis
The daily vaccination rate shows how many vaccinations are given each day. It is calculated using the difference between consecutive days. The data is organized by country and date for easy understanding.

The results show that countries like China, India, and the USA have the highest daily vaccination rates, meaning they are vaccinating people faster. Some countries have lower rates, showing slower progress. This helps compare how quickly different countries are carrying out vaccination programs.

```python
subset = df[df['country'].isin(['India', 'United States'])]

data = subset[['country', 'date', 'total_vaccinations']]

dropna_data = data.dropna()

ffill_data = data.ffill()

print("Before handling missing values:\n", data.describe())

print("\nAfter dropna():\n", dropna_data.describe())

print("\nAfter ffill():\n", ffill_data.describe())
```

### Analysis
The results show that dropna() reduces the dataset size by removing rows with missing values, leading to loss of information. In contrast, forward-fill (ffill) keeps the dataset size unchanged by filling missing values with previous data. The summary statistics show that dropna() reduces the number of records, while ffill() maintains continuity with only slight changes in values. Therefore, ffill() is more suitable for time-series data like vaccination records.
